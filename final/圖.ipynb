```python
import matplotlib.pyplot as plt

teams = {
    "Manchester United": 20,
    "Liverpool": 19,
    "Arsenal": 13,
    "Everton": 9,
    "Manchester City": 9,
    "Aston Villa": 7,
    "Chelsea": 6,
    "Sunderland": 6,
    "Sheffield Wednesday": 4,
    "Newcastle United": 4,
    "Brighton": 3,
    "Leeds United": 3,
    "Wolves": 3,
    "Huddersfield Town": 3,
    "Derby County": 2,
    "Tottenham Hotspur": 2,
    "Burnley": 2,
    "Bournemouth": 2,
    "Preston": 2,
    "Leicester City": 1,
    "Nottingham Forest": 1,
    "Ipswich": 1,
    "West Bromwich Albion": 1,
    "Sheffield United": 1
}

# Create a bar chart
plt.figure(figsize=(10, 8)) 
plt.bar(teams.keys(), teams.values())

plt.xlabel('Team') # x-axis label
plt.ylabel('Number of Championships') 
plt.title('Number of Championships per Team') 
plt.xticks(rotation=90) 
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt

# 2000
pre_2000_championships = {
    "Manchester United": 13,
    "Liverpool": 18,
    "Arsenal": 11,
    "Everton": 9,
    "Manchester City": 2,
    "Aston Villa": 7,
    "Chelsea": 1,
    "Sunderland": 6,
    "Sheffield Wednesday": 4,
    "Newcastle United": 4,
    "Brighton": 3,
    "Leeds United": 3,
    "Wolves": 3,
    "Huddersfield Town": 3,
    "Derby County": 2,
    "Tottenham Hotspur": 2,
    "Burnley": 2,
    "Bournemouth": 2,
    "Preston": 2,
    "Leicester City": 0,
    "Nottingham Forest": 1,
    "Ipswich": 1,
    "West Bromwich Albion": 1,
    "Sheffield United": 1
}

# Sort the teams by championships won before 2000, from highest to lowest
sorted_pre_2000_championships = dict(sorted(pre_2000_championships.items(), key=lambda item: item[1], reverse=True))

# Create a bar chart for pre 2000 championships
plt.figure(figsize=(12, 8)) # Set the size of the figure
plt.bar(sorted_pre_2000_championships.keys(), sorted_pre_2000_championships.values(), color='red')

plt.xlabel('Team') # x-axis label
plt.ylabel('Number of Championships Before 2000') # y-axis label
plt.title('Number of Championships per Team Before 2000 (Sorted)') # Chart title
plt.xticks(rotation=90) # Rotate the x-axis labels by 90 degrees to prevent overlap
plt.tight_layout() # Automatically adjust subplot parameters to fill the figure area

# Display the chart
plt.show()
```
